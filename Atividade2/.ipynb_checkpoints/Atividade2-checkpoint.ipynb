```python
def mask(imagem_bgr, cor):
    img_hsv = cv2.cvtColor(imagem_bgr, cv2.COLOR_BGR2HSV)
    hsv1, hsv2 = aux.ranges(cor)
    return cv2.inRange(img_hsv, hsv1, hsv2)
```

## Parte 3

Assumindo que a folha se mantém sempre paralela ao plano de imagem da câmera, imprima a distância entre a folha e sua câmera

```python
def auto_canny(image, sigma=0.33):
    v = np.median(image)

    lower = int(max(0, (1.0 - sigma) * v))
    upper = int(min(255, (1.0 + sigma) * v))
    edged = cv2.Canny(image, lower, upper)

    return edged

def dist(c1,c2):
	pixel = ((c1[0]-c2[0])**2 + (c1[1]-c2[1])**2)**0.5
	return round((ho/pixel)*f,2)

while(True):
    ret, frame = cap.read()
    
    mask_ciano = mask(frame, '#0c4370')
    mask_magenta = mask(frame,'#86203d')
    
    masks = cv2.bitwise_or(mask_ciano,mask_magenta)
    selecao = cv2.bitwise_and(frame, frame, mask=masks)
    segmentado_cor = cv2.morphologyEx(masks,cv2.MORPH_CLOSE,np.ones((10, 10)))
    selecao = cv2.bitwise_and(frame, frame, mask=segmentado_cor)
    segmentado_masks = cv2.morphologyEx(masks,cv2.MORPH_CLOSE,np.ones((4, 4)))
    
    selecao_ciano = cv2.bitwise_and(frame, frame, mask=mask_ciano)
    segmentado_ciano = cv2.morphologyEx(mask_ciano,cv2.MORPH_CLOSE,np.ones((10, 10)))
    selecao_ciano = cv2.bitwise_and(frame, frame, mask=segmentado_ciano)
    
    selecao_magenta = cv2.bitwise_and(frame, frame, mask=mask_magenta)
    segmentado_magenta = cv2.morphologyEx(mask_magenta,cv2.MORPH_CLOSE,np.ones((10, 10)))
    selecao_magenta = cv2.bitwise_and(frame, frame, mask=segmentado_magenta)
    
    gray_ciano = cv2.cvtColor(selecao_ciano, cv2.COLOR_BGR2GRAY)
    blur_ciano = cv2.GaussianBlur(gray_ciano,(5,5),0)
    bordas_ciano = auto_canny(blur_ciano)
    
    gray_magenta = cv2.cvtColor(selecao_magenta, cv2.COLOR_BGR2GRAY)
    blur_magenta = cv2.GaussianBlur(gray_magenta,(5,5),0)
    bordas_magenta = auto_canny(blur_magenta)
    
    circles_ciano = []
    circles_magenta = []
    circles_total = []

    circles_ciano = None
    circles_ciano = cv2.HoughCircles(bordas_ciano,cv2.HOUGH_GRADIENT,2,40,param1=50,param2=100,minRadius=10,maxRadius=60)
    
    circles_magenta = None
    circles_magenta = cv2.HoughCircles(bordas_magenta,cv2.HOUGH_GRADIENT,2,40,param1=50,param2=100,minRadius=10,maxRadius=60)
    
    if circles_ciano is not None and circles_magenta is not None:
        circles_ciano = np.uint16(np.around(circles_ciano))
        circles_magenta = np.uint16(np.around(circles_magenta))
        
        max_ciano = 0
        index_ciano = None
        max_magenta = 0
        index_magenta = None
        
        for i in range(0,len(circles_ciano)):
            if circles_ciano[i,0,2]>max_ciano:
                max_ciano = circles_ciano[i,0,2]
                index_ciano = i
                
        for i in range(0,len(circles_magenta)):
            if circles_magenta[i,0,2]>max_magenta:
                max_magenta = circles_magenta[i,0,2]
                index_magenta = i
        
        try:
            circles_total.append(circles_ciano[0,index_ciano])
            circles_total.append(circles_magenta[0,index_magenta])

            c1 = [circles_total[0][0], circles_total[0][1]]
            c2 = [circles_total[1][0], circles_total[1][1]]
            
            distancia = dist(c1,c2)
            font = cv2.FONT_HERSHEY_SIMPLEX
            cv2.putText(selecao,'{}'.format(distancia),(0,50), font, 2,(255,255,255),2,cv2.LINE_AA)
        
        except:
            pass
                
    cv2.imshow("Filtro",selecao)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

## Parte 4

Trace uma linha entre os centros do círculo magenta e do círculo ciano.

Imprima na tela o ângulo entre esta linha e a horizontal

```python
def angulo(c1,c2):
	return round(degrees(atan(abs((c1[1]-c2[1])/(c1[0]-c2[0])))),2)

while(True):
    ret, frame = cap.read()
    masks = cv2.bitwise_or(mask(frame, '#0c4370'),mask(frame,'#86203d'))
    selecao = cv2.bitwise_and(frame, frame, mask=masks)
    segmentado_cor = cv2.morphologyEx(masks,cv2.MORPH_CLOSE,np.ones((10, 10)))
    selecao = cv2.bitwise_and(frame, frame, mask=segmentado_cor)
    segmentado_masks = cv2.morphologyEx(masks,cv2.MORPH_CLOSE,np.ones((4, 4)))
        
    gray = cv2.cvtColor(selecao, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray,(5,5),0)
    bordas = auto_canny(blur)
    
    circles = []

    bordas_color = cv2.cvtColor(bordas, cv2.COLOR_GRAY2BGR)
    circles = None
    circles=cv2.HoughCircles(bordas,cv2.HOUGH_GRADIENT,2,40,param1=50,param2=100,minRadius=25,maxRadius=60)

    if circles is not None:
        circles = np.uint16(np.around(circles))
        for i in circles[0,:]:   
            if len(circles[0,:])==2:
                c1 = circles[0,0]
                c2 = circles[0,1]
                distancia = dist(c1,c2)
                inclinacao = angulo(c1,c2)
                font = cv2.FONT_HERSHEY_SIMPLEX
                if distancia > 5:
                    cv2.putText(selecao,'Distancia:{} cm'.format(distancia),(0,50), font, 2,(255,255,255),2,cv2.LINE_AA)
                cv2.putText(selecao,'Angulo:{} graus'.format(inclinacao),(0,100), font, 2,(255,255,255),2,cv2.LINE_AA)
                cv2.line(selecao,(c1[0],c1[1]),(c2[0],c2[1]),(255,0,0),3)
                
    cv2.imshow("Filtro",selecao)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

## Parte 6

Usando `SIFT`, identifique o escrito *Insper* na folha

```python
insper = cv2.imread('insper.png',0)
insper_copy = insper.copy()

sift = cv2.xfeatures2d.SIFT_create()
kpts = sift.detect(insper_copy)
x = [k.pt[0] for k in kpts]
y = [k.pt[1] for k in kpts]
# s will correspond to the neighborhood area
s = [(k.size/2)**2 * pi for k in kpts]
plt.scatter(x, y, s, c='r', alpha=0.1)
plt.imshow(insper_copy, cmap=cm.gray)
plt.title('SIFT')
```

```python
sift = cv2.xfeatures2d.SIFT_create()
kpts = sift.detect(insper_copy)
x = [k.pt[0] for k in kpts]
y = [k.pt[1] for k in kpts]
s = [(k.size/2)**2 * pi for k in kpts]
plt.scatter(x, y, s, c='r', alpha=0.1)

while(True):
    ret, frame = cap.read()
    MIN_MATCH_COUNT = 10
    img_original = insper_copy  
    img_cena = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY) 
    original_rgb = cv2.cvtColor(img_original, cv2.COLOR_GRAY2RGB)
    cena_rgb = cv2.cvtColor(img_cena, cv2.COLOR_GRAY2RGB)
    out = cena_rgb.copy()
    sift = cv2.xfeatures2d.SIFT_create()
    kp1, des1 = sift.detectAndCompute(img_original ,None)
    kp2, des2 = sift.detectAndCompute(img_cena,None)
    FLANN_INDEX_KDTREE = 0
    index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
    search_params = dict(checks = 50)
    flann = cv2.FlannBasedMatcher(index_params, search_params)
    matches = flann.knnMatch(des1,des2,k=2)
    good = []
    for m,n in matches:
    	if m.distance < 0.7*n.distance:
    		good.append(m)
    try:
        if len(good)>MIN_MATCH_COUNT:
            src_pts = np.float32([ kp1[m.queryIdx].pt for m in good ]).reshape(-1,1,2)
            dst_pts = np.float32([ kp2[m.trainIdx].pt for m in good ]).reshape(-1,1,2)
            M, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC,5.0)
            matchesMask = mask.ravel().tolist()
            h,w = img_original.shape
            # Um retângulo com as dimensões da imagem original
            pts = np.float32([ [0,0],[0,h-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)
            # Transforma os pontos do retângulo para onde estao na imagem destino usando a homografia encontrada
            dst = cv2.perspectiveTransform(pts,M)
            # Desenha um contorno em vermelho ao redor de onde o objeto foi encontrado
            img2b = cv2.polylines(out,[np.int32(dst)],True,(255,0,0),3, cv2.LINE_AA)
        else:
            print("Not enough matches are found - %d/%d" % (len(good),MIN_MATCH_COUNT))
            matchesMask = None
    
    except:
        pass
    
    draw_params = dict(matchColor = (0,255,0), # draw matches in green color
	                   singlePointColor = None,
	                   matchesMask = matchesMask, # draw only inliers
	                   flags = 2)
    cv2.imshow("Insper",out)
    if cv2.waitKey(1) & 0xFF == ord('q'):
    	break
```
